```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(100)
delta = np.random.uniform(-15, 15, 100)
y = 0.5 * x + 2 + delta
```

```python
df = pd.DataFrame({'x': x, 'y': y})
df.to_csv("pca-data.csv", index=False)
```
